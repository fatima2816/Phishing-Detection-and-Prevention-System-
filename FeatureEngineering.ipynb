```python

#Ip address extraction
# return 1 for Phishing
# return 0 for Legitimate
import re
def check_ip(url):
    # Regular expression to match an IP
    ip = re.compile(r'\b(?:\d{1,3}\.){3}\d{1,3}\b')
    domain = re.sub(r'^https?://', '', url)
    domain = re.sub(r'/.*$', '', domain)
    if ip.match(domain):
        return 1
    else:
        return 0

```

```python
def check_Length(url):
  if len(url) < 54:
    leng = 0
  else:
    leng = 1
  return leng
```

```python
from urllib.parse import urlparse

def getDepth(url):
    path = urlparse(url).path
    depth = path.count('/')
    return depth
```

```python
#check domain
def check_domain(url):
  domain = urlparse(url).netloc
  if re.match(r"^www.",domain):
	       domain = domain.replace("www.","")
  return domain
```

```python
#CHECKING REDIRECTION

def redirection(url):
  position = url.rfind('//')
  if position > 6 or position > 7:
      return 1
  else:
      return 0
```

```python
# CHECKING FOR HTTP/HTTPS IN DOMAIN NAME
def httpDomain(url):
  domain = urlparse(url).netloc
  if 'https' in domain:
    return 1
  else:
    return 0
```

```python
# Check if the URL contains the '@' symbol
# return 1 for Phishing
# return 0 for Legitimate
def check_symbol(url):
    symbol = re.compile(r'@')
    if symbol.search(url):
        return 1
    else:
        return 0


# url = "http://graphicriver.net@/search?date=this-month"
# result = check_symbol(url)
# print(result)
```

```python
#check for prefix and suffix
# return 1 for Phishing
# return 0 for Legitimate
def check_prefix(url):
    dash = re.compile(r'-')
    # Extract the domain from the URL
    domain = re.sub(r'^https?://', '', url)
    domain = re.sub(r'/.*$', '', domain)
    if dash.search(domain):
        return 1
    else:
        return 0

# url = "http://graphicriver.net-ad/search?-date=this-month"
# result = check_prefix(url)
# print(result)
```

```python
#check for Age of Domain

from datetime import datetime
import whois

def get_domain_info(domain):
    try:
        domain_info = whois.whois(domain)
        return domain_info
    except Exception as e:
        print(f"Error fetching WHOIS information: {e}")
        return None

def calculate_domain_age(domain_info):
    if domain_info is None:
        return 1
    creation_date = domain_info.creation_date
    expiration_date = domain_info.expiration_date

    # check for lists
    if isinstance(creation_date, list):
        creation_date = creation_date[0]
    if isinstance(expiration_date, list):
        expiration_date = expiration_date[0]

   #check for strings
    if isinstance(creation_date, str):
        try:
            creation_date = datetime.strptime(creation_date, '%Y-%m-%d')
        except:
            return 1
    if isinstance(expiration_date, str):
        try:
            expiration_date = datetime.strptime(expiration_date, '%Y-%m-%d')
        except:
            return 1

    # Handle cases where expiration_date or creation_date is None
    if expiration_date is None or creation_date is None:
        return 1
    if not (isinstance(expiration_date, datetime) and isinstance(creation_date, datetime)):
        return 1

    # Calculate the age of the domain in months
    age_of_domain = abs((expiration_date - creation_date).days) // 30

    # If age is less than 6 months, consider it suspicious (phishing)
    if age_of_domain < 6:
        return 1
    else:
        return 0

#example
# url = "http://olx.in/hi/item/nokia-5800-express-music"
# domain_info = get_domain_info(url)
# result = calculate_domain_age(domain_info)
# print(f"The result for {url} is: {result}")
```

```python
#check for end period of domain > 6 months
from datetime import datetime
import whois


def calculate_domain_end(domain_info):
    if domain_info is None:
        return 1
    expiration_date = domain_info.expiration_date
    if isinstance(expiration_date, list):
        return 1
    if isinstance(expiration_date, str):
        try:
            expiration_date = datetime.strptime(expiration_date, "%Y-%m-%d")
        except:
            return 1

    if expiration_date is None:
        return 1

    today = datetime.now()
    remaining_time = abs((expiration_date - today).days) // 30

    return 1 if remaining_time < 6 else 0

# # Example
# url = "http://olx.in/hi/item/nokia-5800-express-music"
# domain_info = get_domain_info(url)
# result = calculate_domain_end(domain_info)
# print(f"The result for {url} is: {result}")
```

```python
import requests
from bs4 import BeautifulSoup

def websiteTraffic(url,timeout=10):
    # Check if the website is accessible
    try:
        response = requests.get(url,timeout=timeout)
        response.raise_for_status()
    except requests.exceptions.RequestException as e:
        print(f"Error accessing the website: {e}")
        return 1
    except requests.exceptions.Timeout:
        print(f"Error: Timeout exceeded ({timeout} seconds)")
        return 1

    # Check if the URL is in the Alexa Top Sites is not recognized
    #if the domain has no traffic return 1 for phishing

    try:
        alexa_url = "https://www.alexa.com/topsites"
        alexa_response = requests.get(alexa_url)
        alexa_response.raise_for_status()

        soup = BeautifulSoup(alexa_response.text, 'html.parser')
        if soup.find('a', string=url):
            return 0
        return 1
    except requests.exceptions.RequestException as e:
        print(f"Error checking Alexa Top Sites: {e}")

    return 0


url = "http://olx.in/hi/item/nokia-5800-express-music"

result = websiteTraffic(url)
print(f"The result for {url} is: {result}")
```

```python
import requests
import re

def check_iframe(url):
    try:
        response = requests.get(url)
        response.raise_for_status()
        if not response.text:
            return 1
        if re.findall(r"[|]", response.text):
            return 0
        else:
            return 1

    except requests.exceptions.RequestException as e:
        # print(f"Error accessing the website: {e}")
        return 1

# Example usage
# url_to_check = "http://extratorrent.cc"
# result = check_iframe(url_to_check)
# print(f"The result for {url_to_check} is: {result}")
```

```python
#Status Bar
def StatusBar(url):
    try:
        response = requests.get(url)
        response.raise_for_status()
        if not response.text:
            return 1
        if re.findall("", response.text):
            return 0
        else:
            return 1
    except requests.exceptions.RequestException as e:
        # print(f"Error accessing the website: {e}")
        return 1

# url = "https://hubpages.com/signin?explain=follow+Hub"
# result=StatusBar(url)
# print(result)
```

```python
#disable right click
def disablerightClick(response):
  try:
        response = requests.get(url)
        response.raise_for_status()
        if not response.text:
            return 1
        if re.findall(r"event.button ?== ?2", response.text):
            return 0
        else:
            return 1
  except requests.exceptions.RequestException as e:
        # print(f"Error accessing the website: {e}")
        return 1

# url = "https://hubpages.com/signin?explain=follow+Hub"
# result=disablerightClick(url)
# print(result)
```

```python
#website forwarding
def webforward(response):
  try:
        response = requests.get(url)
        response.raise_for_status()
        if not response.text:
            return 1
        if len(response.history) <= 2:
            return 0
        else:
            return 1
  except requests.exceptions.RequestException as e:
        # print(f"Error accessing the website: {e}")
        return 1

url = "docs.google.com"
result=webforward(url)
print(result)
```

```python
#store all the features
#check_label for url whether the url is phishing or legitimate
def extractFeatures(url,check_label):
  features = []

  features.append(check_domain(url))
  features.append(check_ip(url))
  features.append(check_symbol(url))
  features.append(check_Length(url))
  features.append(getDepth(url))
  features.append(redirection(url))
  features.append(httpDomain(url))
  features.append(check_prefix(url))

  domain_info = get_domain_info(url)

  features.append(calculate_domain_age(domain_info))
  features.append(calculate_domain_end(domain_info))

  try:
    response = requests.get(url)
    response.raise_for_status()
  except:
    response = ""
    features.append(websiteTraffic(response))
    features.append(check_iframe(response))
    features.append(StatusBar(response))
    features.append(disablerightClick(response))
    features.append(webforward(response))
    features.append(check_label)

  return features


```
